# Preprocessing -  Categorical variables encoding

```python
cat_encoded = pd.get_dummies(categoricals)
cat_encoded
```

```python
#nominal = one hot encoder (dummies)
#ordinal = ordinal encoder (education example)
```

```python
numericals 
```

```python
final_df = pd.concat([numericals,cat_encoded],axis=1)
```

## X y split

```python
X = final_df.drop(columns='tip')

y = final_df['tip']
```

```python
X.head()
```

## Train test split 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, 
                                                    random_state = 100)
```

```python
X_train
```

## scaling / standardizing / power transforming

```python
np.array
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

##powertransformer response 
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
pt.fit(y_train)
y_train_t = pt.transform(y_train).reshape(-1,1))
y_test_t = pt.transform(np.array(y_test).reshape(-1,1))
```
