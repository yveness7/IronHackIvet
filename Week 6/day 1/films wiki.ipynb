```python
url = 'https://en.wikipedia.org/wiki/2023_in_film'
```

```python
from bs4 import BeautifulSoup
import pprint
import requests
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.select('title')
```

```python
print(soup.prettify())
```

```python
for a in soup.select('a'):
    print(a.get_text())
```

```python
movies = []
for movie in soup.select('a'):
    movies.append(movie.get_text())
    
    
movies  
```

## revenue

```python
import re
gross = []
for g in soup.select("table tbody tr td"):
    if re.match('\$',g.get_text()):
        gross.append(g.get_text())
    
gross
```
