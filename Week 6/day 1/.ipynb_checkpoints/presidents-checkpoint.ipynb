```python
#solution

response = requests.get('https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States')
soup = BeautifulSoup(response.content) 

pres_hrefs = []
for i in soup.select('td b a'):
    pres_hrefs.append(i['href'])
    
pres_hrefs #we get the hrefs
```

```python
#we create the lists

name =[]
born =[]
spouse = []
children =[]

#we will iterate now and create the ref soup 
for href in pres_hrefs:
    response = requests.get(base_url+href)
    soup = BeautifulSoup(response.content)
    name.append(soup.select('div.fn')[0].get_text())
    try: 
        born.append(soup.select('span.bday').get_text()) #in case we get any mistakes
    except:
        born.append('NA')
    try:
        spouse.append(soup.select('td.infobox-data div.marriage-display-ws div a').get_text())
    except:
        spouse.append('NA')
    try: 
        children.append(soup.find('th', string = 'Children').parent.select('td.infobox-data')[0].get_text(strip=True))
    except:                  
        children.append(0)                
                       
                       
                       
                       
                       
```

```python
import pandas as pd
presidents_df = pd.DataFrame({'name':name,'born':born,'spouse':spouse,'children':children})
presidents_df
```
