```python
import spotipy
import pandas as pd
import json
from spotipy.oauth2 import SpotifyClientCredentials


#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

```python
import pprint
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
stoner_rock = get_playlist_tracks('Spotify','37i9dQZF1DXdpVGstUksUC')
```

```python
stoner_rock_ids = []
for i in stoner_rock:
    stoner_rock_ids.append(i['track']['id'])
```

```python
chunk = 100 
stoner_rock_ids_chopped = [stoner_rock_ids[i:i+chunk] for i in range(0,len(stoner_rock_ids),100)]
```

```python
stoner_rock_af = []
for i in stoner_rock_ids_chopped:
    try:
        stoner_rock_af.extend(sp.audio_features(i))
    except:
        pass
stoner_rock_df = pd.DataFrame(stoner_rock_af)
stoner_rock_df
```
