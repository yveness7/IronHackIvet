```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

for feature in numerical_features:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
scaler = MinMaxScaler()
df['varA_scaled'] = scaler.fit_transform(df[['total_claim_amount']])

#check that 'varA' has been rescaled using a distribution plot
sns.displot(df['varA_scaled'], kde=True)
plt.title('Distribution of varA after Min-Max Scaling')
plt.show()
```

```python
scaler = StandardScaler()
df['varB_standardized'] = scaler.fit_transform(df[['customer_lifetime_value']])

# Check the mean and standard deviation of the standardized 'varB'
mean_standardized = df['varB_standardized'].mean()
std_standardized = df['varB_standardized'].std()
print(f'Mean after StandardScaler: {mean_standardized}')
print(f'Std after StandardScaler: {std_standardized}')
```

```python
# Plot the PDF after standardization
sns.displot(df['varB_standardized'], kde=True)
plt.title('PDF of varB after StandardScaler')
plt.show()
```

```python

# Check the mean and standard deviation of the power-transformed 'varB'
mean_transformed = df['varB_power_transformed'].mean()
std_transformed = df['varB_power_transformed'].std()
print(f'Mean after PowerTransformer: {mean_transformed}')
print(f'Std after PowerTransformer: {std_transformed}')

# Plot the PDF after power transformation
sns.displot(df['varB_power_transformed'], kde=True)
plt.title('PDF of varB after PowerTransformer')
plt.show()
```
